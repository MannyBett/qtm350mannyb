```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

primes = []
for i in range(1, 101):
    if is_prime(i):
        primes.append(i)

print(primes)
```

```python
def is_prime(num):
    if num <= 1:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    for i in range(3, int(num**0.5) + 1, 2):
        if num % i == 0:
            return False
    return True

primes = [i for i in range(1, 101) if is_prime(i)]

print(primes)
```
Changes Made:
Improved is_prime Function:

Added a check for num == 2 to handle the smallest prime number.
Added a check for even numbers (num % 2 == 0) to quickly eliminate non-prime even numbers.
Changed the loop to iterate only up to the square root of num (int(num**0.5) + 1) and only check odd numbers (range(3, ..., 2)). This reduces the number of iterations and improves efficiency.
List Comprehension:

Used a list comprehension to generate the list of prime numbers, making the code more concise and readable.

```python
def normalize_data(data):
    min_val = min(data)
    max_val = max(data)
    return [(x - min_val) / (max_val - min_val) for x in data]
```

```R
count_vowels <- function(input_string) {
  vowels <- c('a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U')
  sum(strsplit(input_string, NULL)[[1]] %in% vowels)
}

# Example usage
result <- count_vowels("Hello World")
print(result)  # Output: 3
```
